```python
#Calculate the attrition rate 
total_employees = len(df)
employees_left = len(df[df['Attrition'] == 'Yes'])
attrition_rate = (employees_left / total_employees) * 100

print(f'Attrition Rate: {attrition_rate:.2f}%')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Attrition by age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Attrition by Age')
plt.show()
```

```python
# Distribution of years at the company
plt.figure(figsize=(10, 6))
sns.histplot(df['YearsAtCompany'], bins=30, kde=True)
plt.title('Years at the Company Distribution')
plt.show()
```

```python
# Attrition by years at the company
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Attrition by Years at the Company')
plt.show()
```

```python
# Distribution of income
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyIncome'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
# Attrition by income
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Attrition by Income')
plt.show()
```
